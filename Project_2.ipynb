```python
import pandas as pd
import sqlite3
import os
```

```python
# create the sqlite Database
db = sqlite3.connect("JanuaryGroceries.db")
```

```python
# read CSV into dataframe
df1 = pd.read_csv("Python_Spring2020_Project1.csv")
```

```python
# convert the pandas dataframe to a sqlite table
df1.to_sql("purchases", db, if_exists="replace")
```

```python
def run_query(query):
    return pd.read_sql_query(query,db)

os.getcwd()
```

```python
query = """
SELECT * 
FROM sqlite_master
ORDER BY name;
"""
run_query(query)
```
